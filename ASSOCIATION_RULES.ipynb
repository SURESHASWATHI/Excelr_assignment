# Task
Analyze the "Online retail.xlsx" dataset using association rule mining to discover relationships between products, interpret the findings, and answer interview questions about association rules.

## Load the data

### Subtask:
Load the "Online retail.xlsx" dataset into a pandas DataFrame.


**Reasoning**:
Import pandas and load the Excel file into a DataFrame, then display the head and info to understand the data structure.



```python
import pandas as pd

file_path = 'Online retail.xlsx'
df = pd.read_excel(file_path)

display(df.head())
display(df.info())
```

## Data preprocessing

### Subtask:
Handle missing values, clean the data by removing duplicates and ensuring correct data types, and transform the data into a one-hot encoded format suitable for the Apriori algorithm, where each row represents a transaction and each column represents an item.


**Reasoning**:
Convert the single column of comma-separated items into a list of lists, where each inner list represents a transaction and contains individual items as strings. Then, handle potential missing values or empty transactions by removing them and finally create a one-hot encoded DataFrame using `TransactionEncoder`.



```python
from mlxtend.preprocessing import TransactionEncoder

# Convert the single column into a list of lists
transactions = df.iloc[:, 0].astype(str).str.split(',')

# Handle potential empty transactions or missing values
transactions = [t for t in transactions if t and all(item.strip() for item in t)]

# Create a TransactionEncoder
te = TransactionEncoder()

# Fit and transform the transactions into a one-hot encoded DataFrame
te_ary = te.fit(transactions).transform(transactions)
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)

display(df_encoded.head())
display(df_encoded.info())
```

## Implement apriori algorithm

### Subtask:
Use the `mlxtend` library to apply the Apriori algorithm to the pre-processed data (`df_encoded`) to find frequent itemsets.


**Reasoning**:
Apply the Apriori algorithm to find frequent itemsets based on the pre-processed data.



```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets = apriori(df_encoded, min_support=0.01, use_colnames=True)

display(frequent_itemsets.head())
```

## Generate association rules

### Subtask:
Generate association rules from the frequent itemsets found in the previous step.


**Reasoning**:
Generate association rules from the frequent itemsets using the association_rules function with lift as the metric and a minimum threshold of 1, and then display the first few rows of the resulting rules DataFrame.



```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

display(rules.head())
```

## Filter rules

### Subtask:
Filter the generated rules based on appropriate thresholds for support, confidence, and lift.


**Reasoning**:
Filter the rules DataFrame based on lift and confidence thresholds, then sort by lift and display the head.



```python
# Filter rules by lift
filtered_rules = rules[rules['lift'] > 1.2]

# Further filter by confidence
filtered_rules = filtered_rules[filtered_rules['confidence'] > 0.5]

# Sort rules by lift in descending order
filtered_rules = filtered_rules.sort_values(by='lift', ascending=False)

# Display the head of the filtered and sorted rules
display(filtered_rules.head())
```

## Analyze and interpret rules

### Subtask:
Analyze the filtered rules to identify interesting patterns and interpret them to gain insights into customer purchasing behavior.


**Reasoning**:
Examine the filtered rules to identify interesting patterns and interpret them to gain insights into customer purchasing behavior. This involves looking at the antecedents, consequents, support, confidence, and lift values in the `filtered_rules` DataFrame and discussing the implications for a retail business.



```python
print("Interpreting the Filtered Association Rules:")
print("-" * 40)

for index, rule in filtered_rules.iterrows():
    antecedents = ', '.join(list(rule['antecedents']))
    consequents = ', '.join(list(rule['consequents']))
    support = rule['support']
    confidence = rule['confidence']
    lift = rule['lift']

    print(f"Rule: If a customer buys {antecedents}, they are likely to buy {consequents}")
    print(f"Support: {support:.4f} (This rule appears in {support:.2%} of transactions)")
    print(f"Confidence: {confidence:.4f} (Of all transactions containing {antecedents}, {confidence:.2%} also contain {consequents})")
    print(f"Lift: {lift:.4f} (This rule is {lift:.2f} times more likely than random chance)")
    print("-" * 40)

print("\nKey Findings and Practical Implications:")
print("- Rules with high lift, such as (ground beef, eggs) -> (mineral water) and (ground beef, milk) -> (mineral water), indicate strong and non-random associations.")
print("- These findings suggest potential bundling opportunities (e.g., creating a 'breakfast essentials' bundle with ground beef, eggs, and mineral water) or cross-promotion strategies.")
print("- Retailers could use these insights for optimizing product placement (placing associated items near each other), targeted marketing campaigns, and inventory management to ensure frequently bought together items are well-stocked.")
```
