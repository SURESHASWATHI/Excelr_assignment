```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import numpy as np

# Standardize the features
# We exclude the 'Type' column as it is the target variable
features = df.drop('Type', axis=1)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Apply PCA
pca = PCA()
pca.fit(scaled_features)

# Calculate cumulative explained variance
cumulative_explained_variance = np.cumsum(pca.explained_variance_ratio_)

# Plot the explained variance ratio and cumulative explained variance
plt.figure(figsize=(10, 6))
plt.bar(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_, alpha=0.5, align='center', label='Individual explained variance')
plt.step(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, where='mid', label='Cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal component index')
plt.title('Explained Variance by Principal Component')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
# Determine the optimal number of components (e.g., components explaining 95% variance)
n_components = np.argmax(cumulative_explained_variance >= 0.95) + 1
print(f"Number of components explaining at least 95% variance: {n_components}")

# Transform the data with the chosen number of components
pca = PCA(n_components=n_components)
pca_transformed_data = pca.fit_transform(scaled_features)

# Display the shape of the original and transformed data
print(f"Original data shape: {df.shape}")
print(f"PCA transformed data shape: {pca_transformed_data.shape}")
```
